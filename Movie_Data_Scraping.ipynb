```python
import requests
import pandas as pd
import time

# API details
api_key = '04ce269aaeac8736cf7d39d5ac194937'
base_url = "https://api.themoviedb.org/3/movie/popular"
language = "en-US"
num_pages = 427

# List to store results from all pages
all_results = []

# Loop through each page
for page in range(1, num_pages + 1):
    print(f'Scraping page: {page}')
    url = f"{base_url}?api_key={api_key}&language={language}&page={page}"
    
    while True:
        try:
            response = requests.get(url)
            response.raise_for_status()  # Raise an exception for HTTP errors
            data = response.json()
            
            # Extract the results from the JSON data
            results = data['results']
            all_results.extend(results)
            print(f"Successfully scraped page {page}")
            break  # Exit the loop if successful
            
        except requests.exceptions.HTTPError as http_err:
            print(f"HTTP error occurred: {http_err} while scraping page {page}. Retrying...")
        except requests.exceptions.RequestException as err:
            print(f"Error occurred: {err} while scraping page {page}. Retrying...")
        
        # Optional: wait a bit before retrying
        time.sleep(5)

# Create a DataFrame
df = pd.DataFrame(all_results)

# Print the DataFrame
print(df)

# Save the DataFrame to a CSV file
df.to_csv('movies_dataset2.csv', index=False)
```
